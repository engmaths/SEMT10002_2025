```python
x = 5
def change():
    x = 7
    print("Inside:", x)

change()
print("Outside:", x)
```

```python
x = 10
def modify():
    global x
    x = 20

modify()
print(x)
```

```python
def example():
    message = "Hello"
example()
print(message)
```

```python
def outer():
    x = "outer value"
    def inner():
        x = "inner value"
        print("Inside inner:", x)
    inner()
    print("Inside outer:", x)

outer()
```
