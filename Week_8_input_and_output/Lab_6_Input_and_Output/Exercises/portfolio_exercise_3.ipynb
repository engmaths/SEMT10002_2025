## Robot log file

A robot log file is a record of information that a robot (or its control software) automatically writes during operation. It’s used for debugging, performance analysis, diagnostics, and replaying robot behavior.

A log file will typically include information such as:

- Timestamps – when each event or measurement occurred.

- Sensor data – readings from cameras, distance sensors etc.

- Actuator commands – the control signals sent to motors or joints.

- State information – current position, velocity, battery level, or joint angles.

### Generating a timestamp
`perf_counter` from the `time` module gives the time in seconds since Jan 1, 1970 (the Unix epoch). 
The difference in the value returned by `perf_counter` at two points in the program can be used to measure the time between them. 
```python
from time import perf_counter, sleep

start_time = perf_counter()
sleep(1)
stop_time = perf_counter()
print('Process took', stop_time-start_time, 'seconds')
```
