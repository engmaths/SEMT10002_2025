# Line and Scatter Graphs

The `plot` function can be used to generate line and scatter plots of data series

(The `as` keyword can be used to create a user-defined name for `matplotlib.pyplot`; in this case, `plt`)

```python
import matplotlib.pyplot as plt

x = [-1, 3, 4, 8 , 10]
y = [-1, -2, 5, 9 , 11]

plt.plot(x, y) # scatter, o markers, red

plt.show()
```

__Format string:__ an *optional* way to define basic formatting
 - the colour of the plot (e.g. `r` = red, `k`= black)<br>https://matplotlib.org/2.0.2/api/colors_api.html <br>
 - the style of the markers (e.g. `o` = points, `*` = stars)<br>  https://matplotlib.org/api/markers_api.html<br>
 - the style of the line (e.g. `--` = dashes, `.` = dots) <br>https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html

```python
import matplotlib.pyplot as plt

x = [-1, 3, 4, 8 , 10]
y = [-1, -2, 5, 9 , 11]

plt.plot(x, y, 'ro') # scatter, o markers, red

plt.show()
```

__Comprehension check:__ Use the format string to change the appearance of the plot of y against x. 

For example, change the line colour to green, the markers to star shaped and the line style to dashed. 

The `plot` function also takes input arguments (e.g. `color`, `linewi)

```python
import matplotlib.pyplot as plt

x = [-1, 3, 4, 8 , 10]
y = [-1, -2, 5, 9 , 11]

plt.plot(x, y, color='red', linewidth=2.0) 

# Axis labels
plt.xlabel('x data')
plt.ylabel('y data')

# title
plt.title('Plot of f against x')

plt.show()
```

# Saving plots
Plots can be saved in widely used formats (.png, .pdf etc) specified using the file name extension.

`plt.savefig` **must** appear *before* `plt.show()`!

```python
import matplotlib.pyplot as plt

x = [-1, 3, 4, 8 , 10]
y = [-1, -2, 5, 9 , 11]

plt.plot(x, y, 'k.:') 

# Axis labels
plt.xlabel('x data')
plt.ylabel('y data')

# title
plt.title('Plot of f against x')

plt.show()

plt.savefig('my_plot.png')
```

# Importing and plotting data

In this example, two imported modules are used. 

First the csv module is used to import some data, then the matplotlib module is used to plot it 

```python
import csv
import matplotlib.pyplot as plt

# First import some data
with open('temperature.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)
    
print(data)

# Convert the numerical values to numerical data
times = []
temperatures = []

for time, temperature in zip(data[0][1:], data[1][1:]):
    times.append(time)
    temperatures.append(temperature)

# Create a scatter plot of temperature against time 
plt.plot(times, temperature, 'ko', linewidth = 2, markersize = 10)
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.show()

# Save the figure as a PDF file
plt.savefig('temperature_vs_time.pdf')
```

# Importing and plotting multiple data series

Import the data from `temperature_data.txt` 

Notice that this is a txt file. 

The function `.split()` converts the single string to a list of strings

This is not a CSV file so you will not be able to use the `csv` module to help with importing this data.

```python
import csv
import matplotlib.pyplot as plt

# Import the data
with open('temperature_data.txt') as file:
    data = []
    
    # Convert lines to lists of numerical values only
    for line in file:
        print(line)
        
        # Converts the single string to a list of strings
        data = line.split()
        
        # Remove city name
        values = data[1:]
        
        data.append(values)
        
    # Convert to numerical data 
    numerical_data = []
    for row in data:
        numerical_row = []
        for item in row:
            numerical_row.append(float(item))
        numerical_data.append(numerical_row)
    
# Rename
data = numerical_data

# Plot data        
plt.plot(data[0], data[1], label='London')
plt.plot(data[0], data[2], label='Philadelphia')
plt.plot(data[0], data[3], label='Hong Kong')
plt.legend()
plt.show()        
```
